```python
# List of passengers with random seats
passengers = [{'row': row, 'seat': seat, 'seated': False}  # Added flag "seated"
              for row in range(1, ROWS + 1) 
              for seat in ['A', 'B', 'C', 'D', 'E', 'F']]

# Random boarding order
random.shuffle(passengers)

# Function to generate exponentially distributed time
def generate_exp_time(mean_time):
    return np.random.exponential(mean_time)

# Simulation of the boarding process
def simulate_boarding():
    temp = 0
    total_time = 0  # Total boarding time
    passenger_times = []  # Time for each passenger

    # Occupied seats by rows for Block 1 (A, B, C) and Block 2 (D, E, F)
    rows_occupied_block1 = {row: [] for row in range(1, ROWS + 1)}
    rows_occupied_block2 = {row: [] for row in range(1, ROWS + 1)}
    
    # Groups of passengers for group seating (from different rows)
    group_ready_to_sit = []  # List of passengers ready to sit simultaneously (from different rows)
    group_processed = False  # Flag to track if the group has been processed

    for i, passenger in enumerate(passengers):
        row = passenger['row']
        seat = passenger['seat']
        
        # If the passenger is already seated, skip them
        if passenger['seated']:
            continue
        
        # Time for storing luggage
        bag_storage_time = generate_exp_time(MEAN_BAG_STORAGE_TIME)
        
        # Determine the block (Block 1 or Block 2)
        if seat in ['A', 'B', 'C']:
            block = rows_occupied_block1
            seat_order = seat_order_block1
        else:
            block = rows_occupied_block2
            seat_order = seat_order_block2

        # Current passenger and their seat order
        current_seat_order = seat_order[seat]
        occupied_seats = block[row]
        
        # Apply refined row-wide blocking logic (blocking within one row)
        blocking_time_row = 0
        if seat in ['A', 'B', 'C']:  # Block 1 (A, B, C)
            for other_seat in rows_occupied_block1[row]:
                if other_seat in seat_order:  # Ensure the seat is in the dictionary
                    if seat_order[other_seat] > current_seat_order:  # Block only passengers with larger seat order
                        blocking_time_row += generate_exp_time(MEAN_BLOCK_TIME_PER_PASSENGER)
            # Block Block 2 (D, E, F) until someone from Block 1 finishes
            if len(rows_occupied_block2[row]) > 0:
                blocking_time_row += generate_exp_time(MEAN_BLOCK_TIME_PER_PASSENGER * len(rows_occupied_block2[row]))
        elif seat in ['D', 'E', 'F']:  # Block 2 (F, E, D)
            for other_seat in rows_occupied_block2[row]:
                if other_seat in seat_order:  # Ensure the seat is in the dictionary
                    if seat_order[other_seat] > current_seat_order:  # Block only passengers with larger seat order
                        blocking_time_row += generate_exp_time(MEAN_BLOCK_TIME_PER_PASSENGER)
            # Block Block 1 (A, B, C) until someone from Block 2 finishes
            if len(rows_occupied_block1[row]) > 0:
                blocking_time_row += generate_exp_time(MEAN_BLOCK_TIME_PER_PASSENGER * len(rows_occupied_block1[row]))

        # If passengers are from the same row, they cannot be in the group, reset the list
        if any(p['row'] == row for p in group_ready_to_sit):
            group_ready_to_sit.clear()
            group_processed = False  # Reset flag as a new group is forming

        # Group seating is only possible for passengers from different rows
        if len(group_ready_to_sit) >= 2 and all(p['row'] != row for p in group_ready_to_sit):
            temp += 1
            if not group_processed:  # Check if the group has been processed
                # All passengers in the group sit at the same time
                group_sitting_time = generate_exp_time(GROUP_SIT_TIME)
                total_time += group_sitting_time  # Add group time only once
                
                # Account for blocking and seating time for each passenger in the group
                for p in group_ready_to_sit:
                    # Account for individual blocking and seating time for each passenger
                    if p['seat'] in ['A', 'B', 'C']:
                        block = rows_occupied_block1
                        seat_order = seat_order_block1
                    else:
                        block = rows_occupied_block2
                        seat_order = seat_order_block2

                    # Determine the seat order and blocking within the row for each passenger
                    current_seat_order = seat_order[p['seat']]
                    occupied_seats = block[p['row']]

                    individual_blocking_time = 0
                    if p['seat'] in ['A', 'B', 'C']:
                        for other_seat in rows_occupied_block1[p['row']]:
                            if other_seat in seat_order:  # Ensure the seat is in the dictionary
                                if seat_order[other_seat] > current_seat_order:  # Block only passengers with larger seat order
                                    individual_blocking_time += generate_exp_time(MEAN_BLOCK_TIME_PER_PASSENGER)
                    elif p['seat'] in ['D', 'E', 'F']:
                        for other_seat in rows_occupied_block2[p['row']]:
                            if other_seat in seat_order:  # Ensure the seat is in the dictionary
                                if seat_order[other_seat] > current_seat_order:  # Block only passengers with larger seat order
                                    individual_blocking_time += generate_exp_time(MEAN_BLOCK_TIME_PER_PASSENGER)

                    # Add individual time for each group member
                    individual_time = individual_blocking_time + bag_storage_time
                    passenger_times.append(individual_time + group_sitting_time)  # For each group member
                    total_time += individual_time
                group_ready_to_sit.clear()  # Clear the list
                group_processed = True  # Mark the group as processed
        else:
            # If the passenger is not in the group, calculate their individual boarding time
            total_passenger_time = bag_storage_time + blocking_time_row
            total_time += total_passenger_time
            passenger_times.append(total_passenger_time)

        # Add the passenger to the group seating list if they are not from the same row
        group_ready_to_sit.append(passenger)

        # Add the passenger to the list of occupied seats
        block[row].append(seat)
    
    return total_time, passenger_times

# Running the simulation
result, times = simulate_boarding()
print(f"Total boarding time: {result:.2f} minutes")

# Running 100 simulations to calculate average boarding time and confidence interval
num_simulations = 100
boarding_times = []

for _ in range(num_simulations):
    total_time, _ = simulate_boarding()
    boarding_times.append(total_time)

# Analyzing the results
mean_boarding_time = np.mean(boarding_times)
std_boarding_time = np.std(boarding_times)
confidence_interval = 1.96 * std_boarding_time / np.sqrt(num_simulations)

# Outputting the average boarding time and confidence interval
print(f"Average boarding time: {mean_boarding_time:.2f} minutes")
print(f"Confidence interval (95%): from {mean_boarding_time - confidence_interval:.2f} to {mean_boarding_time + confidence_interval:.2f} minutes")


# PDF (Probability Density Function) using Kernel Density Estimation (KDE)
kde = gaussian_kde(boarding_times)
x_values = np.linspace(min(boarding_times), max(boarding_times), 100)
pdf_values = kde(x_values)


# Plotting the histogram, PDF, and CDF
plt.figure(figsize=(10, 6))

# Plotting the histogram of boarding times from 100 simulations
plt.hist(boarding_times, bins=30, color='blue', alpha=0.7, label="Histogram", density=True)

# Plotting the PDF on the same figure
plt.plot(x_values, pdf_values, label="PDF", color='red', linewidth=2)
plt.title("Distribution of Passenger Boarding Times (100 Simulations)")
plt.xlabel("Time (minutes)")
plt.ylabel("Density")
plt.grid(True)
plt.legend()
plt.show()

# CDF (Cumulative Distribution Function)
probabilities = np.arange(1, len(boarding_times) + 1) / len(boarding_times)

# Plotting the CDF (Cumulative Distribution Function)
plt.figure(figsize=(10, 6))
plt.plot(boarding_times, probabilities, label='CDF', color='purple')
plt.title("Cumulative Distribution of Boarding Times (100 Simulations)")
plt.xlabel("Boarding Time (minutes)")
plt.ylabel("Probability")
plt.grid(True)
plt.legend()
plt.show()

```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Constants
ROWS = 50
SEATS_PER_ROW = 6
MEAN_BAG_STORAGE_TIME = 5  # Average time for storing luggage
MEAN_BLOCK_TIME_PER_PASSENGER = 1  # Average time to bypass a blocking passenger

# Function to generate exponentially distributed time
def generate_exp_time(mean_time):
    return np.random.exponential(mean_time)

# Simulate row-by-row boarding process
def simulate_boarding_row_by_row():
    total_time = 0  # Total boarding time
    passenger_times = []  # Time for each passenger

    # Create dictionaries to track occupied seats for each row in both blocks
    rows_occupied_block1 = {row: [] for row in range(1, ROWS + 1)}
    rows_occupied_block2 = {row: [] for row in range(1, ROWS + 1)}

    # Boarding row-by-row (1 to 50), passengers randomly seated within the current row
    for row in range(1, ROWS + 1):
        # Shuffle seating for passengers within the current row
        passengers_row = [{'row': row, 'seat': seat, 'seated': False}
                          for seat in ['A', 'B', 'C', 'D', 'E', 'F']]
        random.shuffle(passengers_row)  # Shuffle seats randomly within the row

        for passenger in passengers_row:
            seat = passenger['seat']

            # Time for storing luggage
            bag_storage_time = generate_exp_time(MEAN_BAG_STORAGE_TIME)

            # Determine the block (Block 1: A, B, C or Block 2: D, E, F)
            if seat in ['A', 'B', 'C']:
                block = rows_occupied_block1
                occupied_seats = block[row]
            else:
                block = rows_occupied_block2
                occupied_seats = block[row]

            # Calculate blocking time based on already seated passengers in the row
            blocking_time_row = 0
            if len(occupied_seats) > 0:
                # Passengers are blocked only by those with a higher seat order
                for other_seat in occupied_seats:
                    blocking_time_row += generate_exp_time(MEAN_BLOCK_TIME_PER_PASSENGER)

            # Total time for the passenger (luggage storage + blocking time)
            total_passenger_time = bag_storage_time + blocking_time_row
            total_time += total_passenger_time
            passenger_times.append(total_passenger_time)

            # Mark the seat as occupied in the current block
            block[row].append(seat)

    return total_time, passenger_times

# Running the refined row-by-row simulation
result_row_by_row_refined, times_row_by_row_refined = simulate_boarding_row_by_row()
print(f"Total boarding time (refined row-by-row): {result_row_by_row_refined:.2f} minutes")

# Running 100 simulations for refined row-by-row boarding to calculate average boarding time and confidence interval
num_simulations = 100
boarding_times_row_by_row_refined = []

for _ in range(num_simulations):
    total_time, _ = simulate_boarding_row_by_row()
    boarding_times_row_by_row_refined.append(total_time)

# Sorting the boarding times for CDF and PDF calculations
boarding_times_row_by_row_refined = np.sort(boarding_times_row_by_row_refined)

# CDF (Cumulative Distribution Function)
probabilities = np.arange(1, len(boarding_times_row_by_row_refined) + 1) / len(boarding_times_row_by_row_refined)

# PDF (Probability Density Function) using Kernel Density Estimation
kde = gaussian_kde(boarding_times_row_by_row_refined)
x_values = np.linspace(min(boarding_times_row_by_row_refined), max(boarding_times_row_by_row_refined), 1000)
pdf_values = kde(x_values)

# Plotting the refined distribution of passengers and their boarding times for row-by-row boarding
plt.figure(figsize=(10, 6))
plt.hist(boarding_times_row_by_row_refined, bins=30, color='orange', alpha=0.7, label="Histogram", density=True)
plt.plot(x_values, pdf_values, label="PDF", color='blue', linewidth=2)
plt.title(f"Distribution of Passenger Boarding Times (Refined Row-by-Row)\nAverage: {np.mean(boarding_times_row_by_row_refined):.2f} minutes")
plt.xlabel("Time (minutes)")
plt.ylabel("Density")
plt.grid(True)
plt.legend()
plt.show()

# Plotting the CDF (Probability vs Boarding Time)
plt.figure(figsize=(10, 6))
plt.plot(boarding_times_row_by_row_refined, probabilities, label='CDF', color='purple')
plt.title("Cumulative Distribution of Boarding Times (Refined Row-by-Row)")
plt.xlabel("Boarding Time (minutes)")
plt.ylabel("Probability")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def simulate_boarding_classic_steffen():
    total_time = 0  # Total boarding time
    passenger_times = []  # Time for each passenger

    # Steffen method groups (Window -> Middle -> Aisle, alternating rows)
    # Group passengers by seat location (e.g., A/F -> B/E -> C/D)

    # First group: Window seats (A and F)
    passengers_group_1 = [{'row': row, 'seat': 'A', 'seated': False} for row in range(ROWS, 0, -1) if row % 2 == 0] + \
                         [{'row': row, 'seat': 'F', 'seated': False} for row in range(ROWS, 0, -1) if row % 2 != 0]

    # Second group: Middle seats (B and E)
    passengers_group_2 = [{'row': row, 'seat': 'B', 'seated': False} for row in range(ROWS, 0, -1) if row % 2 == 0] + \
                         [{'row': row, 'seat': 'E', 'seated': False} for row in range(ROWS, 0, -1) if row % 2 != 0]

    # Third group: Aisle seats (C and D)
    passengers_group_3 = [{'row': row, 'seat': 'C', 'seated': False} for row in range(ROWS, 0, -1) if row % 2 == 0] + \
                         [{'row': row, 'seat': 'D', 'seated': False} for row in range(ROWS, 0, -1) if row % 2 != 0]

    # Combine the groups into boarding process in correct order
    boarding_groups = [passengers_group_1, passengers_group_2, passengers_group_3]

    # Randomize passengers within each group and board one by one
    for group in boarding_groups:
        random.shuffle(group)  # Shuffle the passengers in the group

        for passenger in group:
            row = passenger['row']
            seat = passenger['seat']

            # Time for storing luggage
            bag_storage_time = generate_exp_time(MEAN_BAG_STORAGE_TIME)

            # Add the passenger's time to the total and list
            total_passenger_time = bag_storage_time
            total_time += total_passenger_time
            passenger_times.append(total_passenger_time)

    return total_time, passenger_times

# Running the refined classic Steffen method simulation
result_classic_steffen, times_classic_steffen = simulate_boarding_classic_steffen()
print(f"Total boarding time (Classic Steffen Method): {result_classic_steffen:.2f} minutes")

# Running 100 simulations for refined decreasing row-by-row boarding to calculate average boarding time and confidence interval
num_simulations = 100
boarding_times_classic_steffen = []

for _ in range(num_simulations):
    total_time, _ = simulate_boarding_classic_steffen()
    boarding_times_classic_steffen.append(total_time)

# Analyzing the results for refined decreasing row-by-row boarding
mean_boarding_time_classic_steffen = np.mean(boarding_times_classic_steffen)
std_boarding_time_classic_steffen = np.std(boarding_times_classic_steffen)
confidence_interval_classic_steffen = 1.98 * std_boarding_time_classic_steffen / np.sqrt(num_simulations)

# Outputting the average boarding time and confidence interval for refined decreasing row-by-row boarding
print(f"Average boarding time (simulate_boarding_classic_steffen): {mean_boarding_time_classic_steffen:.2f} minutes")
print(f"Confidence interval (95%): from {mean_boarding_time_classic_steffen - confidence_interval_classic_steffen:.2f} to {mean_boarding_time_classic_steffen + confidence_interval_classic_steffen:.2f} minutes")

# Sorting the boarding times for the cumulative distribution and PDF calculation
boarding_times_classic_steffen = np.sort(boarding_times_classic_steffen)

# CDF (Cumulative Distribution Function)
probabilities = np.arange(1, len(boarding_times_classic_steffen) + 1) / len(boarding_times_classic_steffen)

# PDF (Probability Density Function) using Kernel Density Estimation (KDE)
kde = gaussian_kde(boarding_times_classic_steffen)
x_values = np.linspace(min(boarding_times_classic_steffen), max(boarding_times_classic_steffen), 100)
pdf_values = kde(x_values)

# Plotting the histogram and the PDF
plt.figure(figsize=(10, 6))
plt.hist(boarding_times_classic_steffen, bins=30, color='purple', alpha=0.7, label="Histogram", density=True)
plt.plot(x_values, pdf_values, label="PDF", color='orange', linewidth=2)
plt.title(f"Distribution of Passenger Boarding Times (Classic Steffen Method)\nAverage: {mean_boarding_time_classic_steffen:.2f} minutes")
plt.xlabel("Boarding Time (minutes)")
plt.ylabel("Density")
plt.grid(True)
plt.legend()
plt.show()

# Plotting the CDF (Probability vs Boarding Time)
plt.figure(figsize=(10, 6))
plt.plot(boarding_times_classic_steffen, probabilities, label='CDF', color='purple')
plt.title("Cumulative Distribution of Boarding Times (Classic Steffen Method)")
plt.xlabel("Boarding Time (minutes)")
plt.ylabel("Probability")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Function to simulate boarding with 4 groups, each with 25 smaller groups of 3 people seated together
def simulate_4_large_groups_boarding(passenger_groups, rows):
    total_time = 0
    passenger_times = []

    # Create dictionaries to track occupied seats for each row in both blocks
    rows_occupied_block1 = {row: [] for row in range(1, rows + 1)}
    rows_occupied_block2 = {row: [] for row in range(1, rows + 1)}
    seat_order_block1 = {'A': 0, 'B': 1, 'C': 2}  # Block 1: A < B < C
    seat_order_block2 = {'F': 0, 'E': 1, 'D': 2}  # Block 2: F < E < D

    row_index = rows
    large_group_counter = 0  # Track how many large groups have been seated

    while large_group_counter < 4 and row_index > 0:
        # Seat passengers in large groups (4 total groups)
        for large_group in passenger_groups:
            if large_group_counter < 4:  # Seat only 4 large groups
                small_group_counter = 0  # Track the 25 smaller groups inside each large group
                for small_group in large_group:
                    if small_group_counter < 25 and row_index > 0:
                        for passenger in small_group:
                            row = row_index
                            seat = passenger['seat']

                            # Time for storing luggage
                            bag_storage_time = generate_exp_time(MEAN_BAG_STORAGE_TIME)

                            # Determine the block (Block 1: A, B, C or Block 2: D, E, F)
                            if seat in ['A', 'B', 'C']:
                                block = rows_occupied_block1
                                seat_order = seat_order_block1
                            else:
                                block = rows_occupied_block2
                                seat_order = seat_order_block2

                            # Current passenger and their seat order
                            current_seat_order = seat_order[seat]
                            occupied_seats = block[row]

                            # Apply blocking logic within the same row
                            blocking_time_row = 0
                            if seat in ['A', 'B', 'C']:  # Block 1 (A, B, C)
                                for other_seat in rows_occupied_block1[row]:
                                    if other_seat in seat_order:  # Ensure the seat is in the dictionary
                                        if seat_order[other_seat] > current_seat_order:  # Block only passengers with larger seat order
                                            blocking_time_row += generate_exp_time(MEAN_BLOCK_TIME_PER_PASSENGER)
                                # Block Block 2 (D, E, F) until someone from Block 1 finishes
                                if len(rows_occupied_block2[row]) > 0:
                                    blocking_time_row += generate_exp_time(MEAN_BLOCK_TIME_PER_PASSENGER * len(rows_occupied_block2[row]))
                            elif seat in ['D', 'E', 'F']:  # Block 2 (F, E, D)
                                for other_seat in rows_occupied_block2[row]:
                                    if other_seat in seat_order:  # Ensure the seat is in the dictionary
                                        if seat_order[other_seat] > current_seat_order:  # Block only passengers with larger seat order
                                            blocking_time_row += generate_exp_time(MEAN_BLOCK_TIME_PER_PASSENGER)
                                # Block Block 1 (A, B, C) until someone from Block 2 finishes
                                if len(rows_occupied_block1[row]) > 0:
                                    blocking_time_row += generate_exp_time(MEAN_BLOCK_TIME_PER_PASSENGER * len(rows_occupied_block1[row]))

                            # Total time for the passenger (luggage storage + blocking time)
                            total_passenger_time = bag_storage_time + blocking_time_row
                            total_time += total_passenger_time
                            passenger_times.append(total_passenger_time)

                            # Mark the seat as occupied in the current block
                            block[row].append(seat)

                        # Move to the next row after seating a small group
                        row_index -= 1
                        small_group_counter += 1

                large_group_counter += 1

    return total_time, passenger_times

# Generate random passenger groups of 3 for 4 large groups (4x25x3 = 300 passengers)
def generate_passenger_groups_for_4_large_groups(rows, seats_per_row, group_size=3):
    seat_letters = ['A', 'B', 'C', 'D', 'E', 'F']
    seating_arrangement = [{'row': (i // seats_per_row) + 1, 'seat': seat_letters[i % seats_per_row]} 
                           for i in range(rows * seats_per_row)]
    random.shuffle(seating_arrangement)  # Shuffle seats for randomness

    # Create 4 large groups, each containing 25 smaller groups of 3 passengers
    large_groups = []
    start_index = 0
    for _ in range(4):
        large_group = []
        for _ in range(25):  # Each large group consists of 25 smaller groups
            small_group = seating_arrangement[start_index:start_index + group_size]
            large_group.append(small_group)
            start_index += group_size
        large_groups.append(large_group)

    return large_groups

# Generate the passenger groups (4 large groups, each with 25 smaller groups of 3 passengers)
passenger_groups_4_large_groups = generate_passenger_groups_for_4_large_groups(ROWS, SEATS_PER_ROW)

# Run the boarding simulation for 4 large groups
result_large_group_boarding, times_large_group_boarding = simulate_4_large_groups_boarding(passenger_groups_4_large_groups, ROWS)

# Running multiple simulations to calculate average boarding time and confidence interval
boarding_times_large_group_boarding = []

for _ in range(num_simulations):
    total_time, _ = simulate_4_large_groups_boarding(passenger_groups_4_large_groups, ROWS)
    boarding_times_large_group_boarding.append(total_time)

# Sorting the boarding times for CDF and PDF calculations
boarding_times_large_group_boarding = np.sort(boarding_times_large_group_boarding)

# CDF (Cumulative Distribution Function)
probabilities_large_group = np.arange(1, len(boarding_times_large_group_boarding) + 1) / len(boarding_times_large_group_boarding)

# PDF (Probability Density Function) using Kernel Density Estimation
kde_large_group = gaussian_kde(boarding_times_large_group_boarding)
x_values_large_group = np.linspace(min(boarding_times_large_group_boarding), max(boarding_times_large_group_boarding), 1000)
pdf_values_large_group = kde_large_group(x_values_large_group)

# Plotting the distribution of passenger boarding times for 4 large groups
plt.figure(figsize=(10, 6))
plt.hist(boarding_times_large_group_boarding, bins=30, color='orange', alpha=0.7, label="Histogram", density=True)
plt.plot(x_values_large_group, pdf_values_large_group, label="PDF", color='blue', linewidth=2)
plt.title(f"Distribution of Passenger Boarding Times (4 Large Groups, Back to Front)\nAverage: {np.mean(boarding_times_large_group_boarding):.2f} minutes")
plt.xlabel("Time (minutes)")
plt.ylabel("Density")
plt.grid(True)
plt.legend()
plt.show()

# Plotting the CDF (Probability vs Boarding Time)
plt.figure(figsize=(10, 6))
plt.plot(boarding_times_large_group_boarding, probabilities_large_group, label='CDF', color='blue')
plt.title("Cumulative Distribution of Boarding Times (4 Large Groups, Back to Front)")
plt.xlabel("Boarding Time (minutes)")
plt.ylabel("Probability")
plt.grid(True)
plt.legend()
plt.show()
```
